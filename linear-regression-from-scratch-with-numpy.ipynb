```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../input/salary/Salary.csv')
data.head()
```

```python
x = data['YearsExperience']
y = data['Salary']

plt.scatter(x,y)
plt.title('Salary vs Years of Experience')
plt.show()
```

```python
#y = mx + b
m = 0 
b = 0

Learning_Rate = .01
epochs = 5000

n = np.float(x.shape[0])
error = []

for i in range(epochs):
    y_hat = m*x + b
    mse = (1/n) * np.sum((y-y_hat)**2)
    error.append(mse)
    
    derivative_m = (-2/n) * np.sum(x*(y - y_hat))
    derivative_b = (-2/n) * np.sum(y - y_hat)
    
    m = m - Learning_Rate * derivative_m
    b = b - Learning_Rate * derivative_b
    
x_line = np.linspace(0,15,100)
y_line = m*x_line + b
plt.figure(figsize=(8,6))

plt.title('Salary vs Years of Experience')
plt.plot(x_line,y_line)

plt.scatter(x,y)
plt.show()


plt.plot(error)
plt.show()
```
